## Importação

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
```

## Importando Base de Dados

```python
df_videos=pd.read_excel('videosYT.xlsx',"videos")
display(df_videos)
```

## Tipos de Gráficos

Documentação : https://seaborn.pydata.org/api.html#distribution-api
<br>



## Criando um gráfico de dispersão:
### Scatterplot
Documentação : https://seaborn.pydata.org/generated/seaborn.scatterplot.html#seaborn.scatterplot

```python
fig = sns.scatterplot(data=df_videos,
                      x="Nº de Views",
                      y="Nº de Likes",
                      hue="Responsável")
```

```python
fig = sns.scatterplot(data=df_videos,
                      x="Nº de Views",
                      y="Nº de Likes",
                      hue="Responsável",
                      style="Responsável",
                    )
```

```python
fig = sns.scatterplot(data=df_videos,
                      x="Nº de Views",
                      y="Nº de Likes",
                      hue="Responsável",
                      style="Responsável",
                      palette=["blue","orange"],
                      size = "Nº Inscrições",
                      legend="full"
                    )

```

## Criando um gráfico de dispersão(relacional):
### relplot
Documentação : https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot


```python
fig = sns.relplot(data=df_videos, x="Nº de Views",y="Nº de Likes",col="Categoria")#kind="line"
fig.set_axis_labels(x_var="N° views", y_var="N° Likes")
fig.set_titles("Categoria:{col_name}")
plt.show(fig)
```

## Criando um gráfico de linha:
### lineplot
usando para o mes e ano<br>
Documentação : https://seaborn.pydata.org/generated/seaborn.lineplot.html#seaborn.lineplot

```python
df_inscritos=pd.read_excel('videosYT.xlsx',"Inscritos")
display(df_inscritos)

graf_linha = sns.lineplot(data = df_inscritos, x="Mês/Ano", y="Inscritos", color = "red" )
graf_linha.set_title("Crescimento Inscritos")
plt.show(graf_linha)
```

## Criando Histogramas:
### distplot
Documentação : https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot


```python
fig = sns.displot(data = df_videos,  x = "Nº de Views", ) #kind= histplot
plt.show(fig)
```

```python
fig = sns.displot(data = df_videos,  x = "Nº de Views", kind='kde') #kind= histplot
plt.show(fig)
```

```python
fig = sns.displot(data = df_videos,  x = "Nº de Views", kind='ecdf') #kind= histplot
plt.show(fig)
```

```python
fig = sns.displot(data = df_videos,  x = "Nº de Likes", hue="Responsável", col="Responsável" )#kind = "kde"
plt.show(fig)
```

```python
fig = sns.pairplot(data=df_videos, hue="Categoria")
plt.show(fig)
```

## Criando uma regressão linear:

Documentação: https://seaborn.pydata.org/tutorial/regression.html#functions-to-draw-linear-regression-models

```python
fig = sns.regplot ( data= df_videos, x = "Nº de Views", y = "Nº de Likes") 
plt.show(fig)
```

```python
fig = sns.lmplot ( data= df_videos, x = "Nº de Views", y = "Nº de Likes", hue= "Responsável", markers = ['o','x']) 
plt.show(fig)
```
